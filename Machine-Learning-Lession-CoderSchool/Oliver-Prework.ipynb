```python
def plot_corr_with_labels(df):
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(20,20))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_corr_with_labels(df)
```

```python
under_45 = df[ df["AGE"] < 45]
over_45 = df[ df["AGE"] >= 45]

print(under_45.size)
print(over_45.size)
```

```python
u45 = under_45.corr()["ADIPOSITY"]
o45 = over_45.corr()["ADIPOSITY"]

for i in range(df.columns.values.size):
    print (u45[i] - o45[i])
```

```python
list(df.columns.values)
```

```python
changes = [];
for i in range(df.columns.values.size):
    changes.append(u45[i] - o45[i])
for i in range(df.columns.values.size):
    print(df.columns.values[i], '\t', changes[i])
```

```python
max_change_i = 0;
for i in range(len(changes)):
    if changes[max_change_i] < changes[i]:
        max_change_i = i;
print(df.columns.values[max_change_i], '\t', changes[max_change_i]) 
```

```python
def plot_changes_with_labels(columns, changes):
    fig, ax = plt.subplots(figsize=(15,1))
    ax.matshow([changes])
    plt.yticks([0], ['changes'])
    plt.xticks(range(len(changes)), columns.values)
plot_changes_with_labels(df.columns, changes)
```
